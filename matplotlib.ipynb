```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,2,1,4,6,4,7,5]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

```python
plt.plot(x,y,color = "Black")
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

```python
plt.plot(x,y,color = "red",linestyle ="dashdot")#gives dashdot line
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

```python
plt.plot(x,y,color = "red",linewidth =3)#gives dashdot line
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

```python
plt.plot(x,y,marker = "*")
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

```python
plt.plot(x,y,marker = "*",markersize = 20)
plt.title("Numbers count frequency")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()# it is used to display the graph
```

Bar Graph

```python
x = ['python','c','c++','java']
y = [87,62,46,80]
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,color='black',width=0.4)
plt.title("Programming languages vs popularity")
plt.xlabel("languages")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.bar(x,y,color='black',width=0.4,align = 'edge')
plt.title("Programming languages vs popularity")
plt.xlabel("languages")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.bar(x,y,color='orange',width=0.4,align = 'edge',edgecolor='r',linewidth=3,linestyle = '-.')
plt.title("Programming languages vs popularity")
plt.xlabel("languages")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.bar(x,y,color='black',width=0.4,alpha = 0.3)
plt.title("Programming languages vs popularity")
plt.xlabel("languages")
plt.ylabel("Popularity")
plt.show()
```

Histogram 

```python
x = np.random.randn(20)
plt.hist(x)
plt.show()
```

```python
plt.hist(x)
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
plt.hist(x,color = 'm',edgecolor = 'black')
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
plt.hist(x,color = 'm',edgecolor = 'black',bins = 15)
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
plt.hist(x,color = 'm',edgecolor = 'black',cumulative = -1)
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
plt.hist(x,color = 'm',histtype='step')
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
plt.hist(x,color = 'm',edgecolor = 'black',orientation='horizontal')
plt.title("Frequent count")
plt.ylabel('count')
plt.xlabel('numbers')
plt.show()
```

```python
x = [10,20,30,25,15]
y = ['c','c++','pyhton','java','r']
```

```python
plt.pie(x)
plt.show()
```

```python
plt.pie(x,labels = y)
plt.show()
```

```python
ex = [0.2,0.0,0.0,0.1,0.0]
plt.pie(x,labels=y,explode = ex)
plt.show()
```

```python
c = ['r','g','b','y','m']
plt.pie(x,labels = y,colors = c)
plt.show()
```

```python
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True)
plt.show()
```

```python
ex = [0.2,0.0,0.0,0.1,0.0]
plt.pie(x,labels=y,explode = ex,radius = 1.5)
plt.show()
```

```python
plt.pie(x,labels = y,labeldistance=1.2)
plt.show()
```
