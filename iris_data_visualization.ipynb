```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = sns.load_dataset("iris")
```

```python
print(x.head())
```

Question 1->Relationship Between Petal Length and Petal Width Across Species

```python
sns.scatterplot(data=x, x ='petal_length', y ='petal_width', hue ='species')
plt.title("Petal Length vs Petal Width")
plt.show()
```

Findings from visual 1.


From the above visual i can give following insights.
1. The petal length and petal width of setosa species is very small comparatively to others species.
2. Petal length and petal width of virginica is larger among all species.
3. Petal length and petal width if versicolor is larger than setosa and smaller than virginica.

Question 2->Trend of Sepal Length by Species - Show how average sepal_length varies across species

```python
sns.relplot(data=x,x="species",y="sepal_length",kind="line",errorbar="sd",marker="o")
plt.title("Average Sepal Length by Species")
plt.show()
```

Findings from visual 1.

1. Virginica has the highest average sepal length.

2. Setosa has the lowest.

Question 3->Comparison of Sepal Width Across Species - Create a bar plot to compare mean sepal_width for each species.

```python
sns.barplot(data=x, x='species', y='sepal_width')
plt.title("Mean Sepal Width by Species (barplot)")
plt.show()
```

Findings from visual 3.

1. Setosa tends to have higher average sepal width.

2. Virginica and Versicolor have similar, slightly lower values.

3. Less useful for species separation.
